```python
import matplotlib.pyplot as plt 
import pickle 
import numpy as np 
```

```python
with open("DDPG1.pkl", "rb") as f: 
    data_w = pickle.load(f)

with open("DDPG2.pkl", "rb") as f: 
    data_norm = pickle.load(f)

with open("DDPG3.pkl", "rb") as f: 
    data_batchnorm = pickle.load(f)
```

```python
plt.figure(figsize=(10, 7))

for i in range(2): 
    if i == 0: 
        plt.plot(data_w[i+1]["rewards"], "k", label="power in [W]", linewidth=2)
    else: 
        plt.plot(data_w[i+1]["rewards"], "k", linewidth=2)
for i in range(2): 
    if i == 0: 
        plt.plot(data_norm[i+1]["rewards"], "b", label="normalized power", linewidth=2)
    else: 
        plt.plot(data_norm[i+1]["rewards"], "b", linewidth=2)
for i in range(2): 
    if i == 0: 
        plt.plot(data_batchnorm[i+1]["rewards"], "r", label="batch normalization layer", linewidth=2)
    else: 
        plt.plot(data_batchnorm[i+1]["rewards"], "r", linewidth=2)

plt.xlabel("episodes", fontsize=14)
plt.ylabel("episodic rewards", fontsize=14)
plt.legend(fontsize=14)
plt.grid()
plt.savefig("results.png")
```
