```python
import matplotlib.pyplot as plt 
import pickle 
```

```python
with open("DDPG.pkl", "rb") as f: 
    data_DDPG = pickle.load(f)

with open("DDQN.pkl", "rb") as f: 
    data_DDQN = pickle.load(f)
    
data_DDQN
```

```python
plt.figure(figsize=(10, 7))

for i, data_DDPG in data_DDPG.items(): 
#     print(data_DDPG)
    plt.plot(data_DDPG["rewards"], 'r', label="episodic rewards DDPG", linewidth=2, )

for i, data_DDQN in data_DDQN.items(): 
    plt.plot(data_DDQN["rewards"], 'b', label="episodic rewards DDPG", linewidth=2, ) 



plt.xlabel("episodes", fontsize=14)
plt.ylabel("episodic rewards", fontsize=14)
plt.legend(fontsize=14)
plt.grid()
plt.savefig("Comparison between DDQN and DDPG.png")
```
