```python
import matplotlib.pyplot as plt 
import pickle 
```

```python
with open("results_scaling_400.pkl", "rb") as f: 
    data_scale = pickle.load(f)[10]

with open("results_notScaling_400.pkl", "rb") as f: 
    data_notScale = pickle.load(f)[10]
    
# data_scale 
```

```python
plt.figure(figsize=(14, 10))
plt.plot(data_scale["rewards"], label="Power in [kW]", linewidth=2)
plt.plot(data_notScale["rewards"], label="Power in [W]", linewidth=2)
plt.xlabel("episodes", fontsize=16)
plt.ylabel("episodic rewards", fontsize=16)
plt.legend(fontsize=14)
plt.xlim([0, 130])
plt.grid()
plt.savefig("results.png")
```
